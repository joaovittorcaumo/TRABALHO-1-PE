```python
data = {'NUMEROS':[15938,18919,25143,16077,19601,16510,21582,325689,17053,17629,18174,28901]}
df = pd.DataFrame(data) # CRIANDO O DATA FRAME PARA A QUESTÃO 4
# 4 A
df.mean() # A RECEITA MEDIA É DE 45,101.33
```

```python
# SETANDO O DATA FRAME PARA A QUESTÃO 6
df = pd.DataFrame({
    'X': [133, 99, 70, 160, 190, 153, 217],
    'Y': [77, 33, 20, 30, 42, 37, 20]
})
df
```

```python
# 6 A
df.mean() # O INVESTIMENTO MEDIO DO GRUPO X É DE 146 E O DO GRUPO Y É 37
```

```python
# 6 B
df.median() # A MEDIANA DOS INVESTIMENTOS DO GRUPO X É DE 153 E DO GRUPO Y É 33
```

```python
# 6 C
# NENHUM DOS GRUPOS APRESENTA MODA, POÍS NENHUM DOS NUMEROS SE REPETEM NA TABELA.
```

```python
# 6 D
df.sort_values(by=['Y'])  # A AMPLITUDE DE X E DE 147 (217-70)
                          # A AMPLITUDE DE Y E DE 57  (77-20)
```

```python
# 6 E
df.std() # O DESVIO PADRÃO AMOSTRAL DO GRUPO X É DE 50.62 E DO GRUPO Y 19.44
```

```python
# 6 F
# CALCULADO A AMPLITUDE INTERQUATILICA DE X
q3, q1 = np.percentile(df['X'], [75,25])
iqr = q3 - q1 
iqr
```

```python
# CALCULADO A AMPLITUDE INTERQUATILICA DE Y
q3, q1 = np.percentile(df['Y'], [75,25])
iqr = q3 - q1 
iqr
# RESUMO: A AMPLITUDE INTERQUATILICA DE X E: 59 E A DE Y E: 14.5
```
